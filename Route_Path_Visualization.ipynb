```python
import networkx as nx
import matplotlib.pyplot as plt
import random

G = nx.complete_graph(10)
route_idx = [0, 2, 4, 6, 8]

pos = {i: (random.random(), random.random()) for i in range(len(G))}

plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')

nx.draw_networkx_edges(G, pos, alpha=0.5)

route_edges = [(route_idx[i], route_idx[i+1]) for i in range(len(route_idx)-1)]
nx.draw_networkx_edges(G, pos, edgelist=route_edges, edge_color='red', width=2)

nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

for i, node in enumerate(route_idx):
    x, y = pos[node]
    plt.text(x, y + 0.05, str(i+1), fontsize=12, ha='center', color='black')

plt.title('Route Path Visualization', fontsize=16)

plt.axis('off')
plt.show()
```
